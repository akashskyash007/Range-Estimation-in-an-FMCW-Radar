```python
#Importing Necessary Packages
import numpy as np
from matplotlib import pyplot as plt
import time
from IPython.display import clear_output
```

```python
#Function to Read File
def read_data(file_name):
    f = open(file_name, 'r')
    data = np.fromfile(f, dtype = np.int16)
    return data
```

```python
#Function to prerocess the File Data
def preprocess(data):
    fast_time_samples = 512
    bits_per_sample = 16
    num_receivers = 4
    is_real = 0
    if (bits_per_sample != 16):
        max_value = pow(2, bits_per_sample) - 1
        data = np.where(data > max_value, data - (max_value + 1), data)
    total_samples = data.shape[0]
    if (is_real):
        num_chirps = int(total_samples/(num_receivers * fast_time_samples))
        data_new = np.reshape(data, (fast_time_samples * num_receivers, num_chirps))
        data_new = data_new.T
        preprocessed_data = np.zeros((num_receivers, fast_time_samples * num_chirps))
        for curr_receiver in range(num_receivers):
            for curr_chirp in range(num_chirps):
                preprocessed_data[curr_receiver, curr_chirp * fast_time_samples: (curr_chirp + 1) * fast_time_samples] = data_new[curr_chirp, curr_receiver * fast_time_samples:(curr_receiver + 1) * fast_time_samples]
        return preprocessed_data
    else:
        num_chirps = int(total_samples/(2 * fast_time_samples * num_receivers))
        data_new = np.zeros(int(total_samples/2), dtype = 'complex')
        counter = 0
        for i in range(0, total_samples - 1, 4):
            data_new[counter] = data[i] + 1j * data[i+2]
            data_new[counter+1] = data[i+1] + 1j * data[i+3]
            counter = counter + 2
        data_new = np.reshape(data_new, (fast_time_samples * num_receivers, num_chirps), order = "F")
        data_new = data_new.T
        preprocessed_data = np.zeros((num_receivers, fast_time_samples * num_chirps), dtype = 'complex')
        for curr_receiver in range(num_receivers):
            for curr_chirp in range(num_chirps):
                preprocessed_data[curr_receiver, curr_chirp * fast_time_samples: (curr_chirp + 1) * fast_time_samples] = data_new[curr_chirp, curr_receiver * fast_time_samples:(curr_receiver + 1) * fast_time_samples]
        return preprocessed_data
```

```python
#Calling Functions to Read File and Preprocess Data
file_name = 'sr_car.bin'
data = read_data(file_name)
print(len(data))
preprocessed_data = preprocess(data)
```

```python
#Separating data for each of the four receivers
receiver1_data = preprocessed_data[0,:]
receiver2_data = preprocessed_data[1,:]
receiver3_data = preprocessed_data[2,:]
receiver4_data = preprocessed_data[3,:]
```

```python
#Define Constants
c = 3e8
fast_time_sampling_rate = 5e6
fast_time_samples = 512
freq_slope = 8e12
idle_time = 3e-6
ramp_end_time = 56e-6
carrier_frequency = 77e9
start_frequency = carrier_frequency
num_receivers = preprocessed_data.shape[0]
frames = 100
s = int(receiver1_data.shape[0]/frames)
m_chirps = int(s/fast_time_samples)
m_chirps
```

```python
#Calculation of Range Resolution and Doppler Resolution
range_res = (c * fast_time_sampling_rate)/(2 * freq_slope * fast_time_samples)
print(range_res)
doppler_res = (c)/(2 * start_frequency * (ramp_end_time + idle_time) * m_chirps)
print(doppler_res)
```

```python
#Reshaping receiver data in accordance with each row representing a frame
receiver1_data_frame = np.reshape(receiver1_data, (s, frames), order = "F")
receiver2_data_frame = np.reshape(receiver2_data, (s, frames), order = "F")
receiver3_data_frame = np.reshape(receiver3_data, (s, frames), order = "F")
receiver4_data_frame = np.reshape(receiver4_data, (s, frames), order = "F")
receiver1_data_frame.shape
```

```python
#Writing Algorithm for Range Estimation
for curr_frame in range(frames):
    plt.figure(figsize = (15, 7))
    clear_output(wait = True)
    curr_frame_data = receiver1_data_frame[:, curr_frame]
    curr_frame_data = np.reshape(curr_frame_data, (fast_time_samples, m_chirps), order = "F")
    curr_frame_data_odd_chirps = curr_frame_data[:, 1:m_chirps:2]
    range_estimate = 20*np.log10(np.abs(np.fft.fft(curr_frame_data_odd_chirps.T).T))
    pulse_integrated_range_estimate = np.sqrt(np.sum(np.power(range_estimate, 2), axis = 1))
    range_values = np.arange(1, pulse_integrated_range_estimate.shape[0] + 1) * range_res
    plt.plot(range_values, pulse_integrated_range_estimate)
    plt.xlabel('Range in meters')
    plt.ylabel('Likelihood of Object Being Present')
    plt.title('Frame Number : ' + str(curr_frame + 1))
    plt.show()
    time.sleep(1)
```
